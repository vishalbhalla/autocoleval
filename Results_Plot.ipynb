## Plot Table 5. Candidate Ranking AMs Comparison

```python
import os
import numpy as np

import matplotlib as mpl 
mpl.use('TkAgg') 
import matplotlib.pyplot as plt

%matplotlib inline

import pandas as pd
```

```python
# Change path to the folder where the "Elia_CollocationMeasures_Comparison.csv" file (provided with the code) is located.
plotDataFolder = './'
```

```python
xlabel = 'Association Measures (AMs)'
fields = ['Collocation Measure', 'Top 4,626_R', 'Top 14,550_R', 'Top 29,100_R', 'Top 14,550_P', 'Top 29,100_P']
df = pd.read_csv(os.path.join(plotDataFolder, 'Elia_CollocationMeasures_Comparison.csv'), skipinitialspace=True, usecols=fields)
fig = plt.figure()
# legbox = plt.subplot(111)

ax = df[['Top 4,626_R', 'Top 14,550_R', 'Top 29,100_R', 'Top 14,550_P', 'Top 29,100_P']].plot(style=['-','-.','--.','--',':'],)
ax.set(xlabel=xlabel, ylabel='Precision or Recall %')

xi = ['Delta P', 'MI', 'MI2', 'MI3', 'Log Dice', 'Salience', 'FTW', 'Frequency', 'Log Likelihood', 'T-score']

x = df['Collocation Measure']
# create an index for each tick position
x=range(len(xi))
plt.xticks(x, xi, rotation=40)

plt.legend([ax], loc='upper left', bbox_to_anchor=(0.4, -0.4), fancybox=True, shadow=True, 
           labels=['Top 4,626 Recall=Precision', 'Top 14,550 Recall', 'Top 29,100 Recall', 'Top 14,550 Precision', 'Top 29,100 Precision'])

fig.autofmt_xdate()
plt.show()
```

```python
xlabel = 'Association Measures (AMs)'
fields = ['Collocation Measure', 'Top 4,626_R', 'Top 14,550_R', 'Top 29,100_R', 'Top 14,550_P', 'Top 29,100_P']
df = pd.read_csv(os.path.join(plotDataFolder, 'Elia_CollocationMeasures_Comparison.csv'), skipinitialspace=True, usecols=fields)
fig = plt.figure()

from cycler import cycler
bar_cycle = (cycler('hatch', ['///', '--', '...', 'xxx', '\\\\']) * cycler('color', 'w')*cycler('zorder', [10]))
patterns = bar_cycle()

ax = df.plot(kind='bar', width = 0.8, style=patterns,)
ax.set(xlabel=xlabel, ylabel='Precision or Recall %')


xi = ['Delta P', 'MI', 'MI2', 'MI3', 'Log Dice', 'Salience', 'FTW', 'Frequency', 'Log Likelihood', 'T-score']

x = df['Collocation Measure']
# create an index for each tick position
x=range(len(xi))
plt.xticks(x, xi, rotation=40)

plt.legend([ax], loc='upper left', bbox_to_anchor=(0.4, -0.4), fancybox=True, shadow=True, 
           labels=['Top 4,626 Recall=Precision', 'Top 14,550 Recall', 'Top 29,100 Recall', 'Top 14,550 Precision', 'Top 29,100 Precision'])

fig.autofmt_xdate()
plt.show()
```

```python
xlabel = 'Association Measures (AMs)'
fields = ['Collocation Measure', 'Top 4,626_R', 'Top 14,550_R', 'Top 29,100_R', 'Top 14,550_P', 'Top 29,100_P']
df = pd.read_csv(os.path.join(plotDataFolder, 'Elia_CollocationMeasures_Comparison.csv'), skipinitialspace=True, usecols=fields)
fig = plt.figure()

# plt.rcParams["figure.figsize"] = (20,5)
plt.rc('figure', figsize=(15, 6))

ax = fig.add_subplot(111)
ax.set(xlabel=xlabel, ylabel='Precision or Recall %')

xi = ['Delta P', 'MI', 'MI2', 'MI3', 'Log Dice', 'Salience', 'FTW', 'Frequency', 'Log Likelihood', 'T-score']

patterns = ['', '++', '\\\\', 'ooo', '--', 'xxx']
# colors = ['', 'blue', 'orange', 'green', 'red', 'purple']
colors = ['', 'lightskyblue', 'lemonchiffon', 'lightgreen', 'lightsalmon', 'plum']

k=0
for i in range(10):
    for j in range(1, len(patterns)):
        pos = k + j -3
        #ax.bar(k+j-3, height=df.iloc[i, j], width=0.8, color='white', edgecolor='black', hatch=patterns[j], align='center')
        ax.bar(pos, height=df.iloc[i, j], width=1, color=colors[j], edgecolor='black', hatch=patterns[j], align='center')
    k=k+7
    
plt.xticks(np.arange(0, 70, 7), xi, rotation=0)

plt.legend([ax], loc='upper left', bbox_to_anchor=(1.05, 1.05), title="Elia Test Sample", fancybox=True, shadow=True, 
           labels=['Top 4,626 Recall=Precision', 'Top 14,550 Recall', 'Top 29,100 Recall', 'Top 14,550 Precision', 'Top 29,100 Precision'])
plt.show()
```

```python
xlabel = 'Association Measures (AMs)'
fields = ['Collocation Measure', 'Top 4,626_R', 'Top 14,550_R', 'Top 29,100_R', 'Top 14,550_P', 'Top 29,100_P']
df = pd.read_csv(os.path.join(plotDataFolder, 'Elia_CollocationMeasures_Comparison.csv'), skipinitialspace=True, usecols=fields)
fig = plt.figure()

plt.rcParams["figure.figsize"] = (20,5)
# plt.rc('figure', figsize=(15, 6))

ax = fig.add_subplot(111)
ax.set(xlabel=xlabel, ylabel='Precision or Recall %')

xi = ['Top 4,626 Recall=Precision', 'Top 14,550 Recall', 'Top 29,100 Recall', 'Top 14,550 Precision', 'Top 29,100 Precision']

patterns = ['', '+', '\\', 'O', '-', 'X', '//', '+++', 'ooo', '|||', 'xxx']
# colors = ['', 'blue', 'orange', 'green', 'red', 'purple']
colors = ['lightskyblue', 'lemonchiffon', 'lightgreen', 'lightsalmon', 'plum', 'khaki', 'sandybrown', 'lightseagreen', 'mintcream', 'burlywood']

k=0
for i in range(1, 6):
    for j in range(0, len(patterns)-1):
        pos = k + j -5
        ax.bar(pos, height=df.iloc[j, i], width=1, color=colors[j], edgecolor='black', hatch=patterns[j], align='center')
    k=k+12
    
plt.xticks(np.arange(0, 70, 12), xi, rotation=0)

plt.legend([ax], loc='upper left', bbox_to_anchor=(1.05, 1.05), title="Elia Test Sample", fancybox=True, shadow=True, 
           labels=['Delta P', 'MI', 'MI2', 'MI3', 'Log Dice', 'Salience', 'FTW', 'Frequency', 'Log Likelihood', 'T-score'])
plt.show()
```
